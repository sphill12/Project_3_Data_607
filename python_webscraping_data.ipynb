```python
import investpy
import investiny
import pandas
import numpy
import yfinance as yf
import yahoofinance
import os
```

```python
ticker_list = ("ADA",  "ALGO", "APT21794",  "AVAX", "BCH",  "BNB",  "CCD",  "CHZ",  "DOT",  "EGLD", "ERG",  "ETH",  "FLOW", "FTM",  "FUSE", "HBAR", "KAS",  "KDA",  "LYX",  "RXD",  "SEI",  "SOL",  "SOUL","SUI20947",  "TRAC", "XDC",  "XEC",  "XRD",  "XTZ")
new_list = list()
for i in ticker_list:
    i += "-USD"
    new_list.append(i)
```

```python
cryptos = " ".join(new_list)
```

```python
tickers = yf.Tickers(cryptos)
new_data = tickers.history("max")
```

```python
new_data
```

```python
yf.Ticker("MSFT").info["marketCap"]
```

```python
full_list = pandas.DataFrame()
for i in new_list:
    temporary = yf.Ticker(i).history("max")
    temporary["marketcap"] = yf.Ticker(i).info["marketCap"]
    temporary["coin"] = i
    full_list = full_list.append(temporary)
    
```

```python
full_list
```

```python
full_list = full_list.drop(['Dividends',"Stock Splits"], axis = 1)
full_list = full_list.reset_index()
```

```python
full_list = full_list.rename(columns = {"Date":"date","Open":"open", "High":"high","Low":"low","Close":"close","Volume":"volume"})
```

```python
full_list["coin"] = full_list["coin"].str[:-4]
```

```python
full_list
```

```python
replacements_layer_one = {"APT21794":"APT","SUI20947":"SUI"}
```

```python
full_list["coin"] = full_list["coin"].replace(replacements_layer_one)
```

```python
full_list.to_csv("layer_one_daily_data.csv")
```
